```python
#using edit distance

from nltk.corpus import words
from nltk.metrics.distance import edit_distance
correct_words = words.words()
incorrect_words = ["happyi" ,"azmaing" , "intelligent"]
for word in incorrect_words:
    temp =[(edit_distance(word , w) , w)
           for w in correct_words if w[0]== word[0]]
    temp.sort()
    print(temp)
    
```

```python

#using jackcard distance
from nltk.util import ngrams

from nltk.corpus import words
from nltk.metrics.distance import jaccard_distance
correct_words = words.words()
incorrect_words = ["happyi" ,"azmaing" , "intelligent"]
for word in incorrect_words:
    temp =[(jaccard_distance(set(ngrams(word , 2)) ,set(ngrams(w ,2))) ,w)
           for w in correct_words if w[0]== word[0]   ]
    temp.sort()
    print(temp)
```

```python
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

sentence1 =" I love reading books on coding".split()
sentence2 = "the table was already booked by someone by sentence else".split()
wsd1=  lesk(sentence1 , "book")
print(wsd1 .definition())
wsd2 = lesk(sentence2,"book")
print(wsd2 .definition())
```
